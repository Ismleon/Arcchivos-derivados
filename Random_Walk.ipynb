#PRÁCTICA 2 :RANDOM WALK
#Random walk en 1 dimensión
Supongamos que soy un inversor y quiero adquirir acciones de una determinada compañía, pongamos de ejemplo el banco Santander. Actualmente las acción del banco Santander cotizan en torno a 3€. Como he explicado en el blog, el random walk se basa en una evolución aleatoria. Entonces hagamos nuestro random walk con las acciones del banco Santander partiendo de un valor inicial de 3€/acción y se revalorizan un 1,5% si sube o baja el valor de cotización. Realizaremos una simulación a 1000 días. Estudiada la coyuntura económica a la que se enfrenta el banco Santander, conocemos la probabildiad de crecer y decrecer siendo respectivamente del 50%

```python
import random
import matplotlib.pyplot as mat
random.seed()
precio_0 = 3     # inicializamos el valor de la cotización
n = 1000         # establecemos el número de sesiones cotizadas
factor = 1.015    # establecemos el factor, siendo el incremento del 1,5% el factor será 1.015
prob = .5           # probabilidad de subida. Si es 0.5 la probabilidad de que el precio suba es del 50%
valoracion = [precio_0] # inicializamos la valoración de precios, comenzamos con la primera cotización y creamos un bucle para el resto de valores hasta 1000 valoraciones.
for t in range(1,n):
    if random.random() < prob:
        precio_0 *= factor
    else:
        precio_0 /= factor
    valoracion.append(precio_0)
mat.plot(range(n),valoracion)
mat.show()
```

A continuación veremos como se realiza el Random Walk en 2 dimensiones.También es conocido como el paseo del borracho.A diferencia de realizarlo en 1 dimensión, con 2 dimensiones se utiliza el plano cartesiano partiendo el punto (0,0).

```python
#importamos primero las librerias necesarias para realizar el random walk.  
import random  #número aleatorio
import math  #Para realizar los cálculos matemáticos relacionados con los ángulos.
import matplotlib.pyplot as mat  #Libreria empleada para la representación de gráficos
random.seed()  
n=30 # número de pasos  
r=1 # longitud del paso que se da en la simulación del paseo 
x=y=0 # punto de partida (x,y)=(0,0)  
origenx=[x]  
origeny=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 # el ángulo está medido en radianes y es plenamente aleatorio. 
  #De esta manera se consigue que la dirección del paso se desconozca hacia donde va a ir.  
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  origenx.append(x)  
  origeny.append(y)  
mat.plot(origenx,origeny)  
mat.show()
```
